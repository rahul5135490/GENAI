```python
from transformers import pipeline

# Load translation pipeline
translator = pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi")

# Input sentence
sentence = "How are you doing today?"

# Translate
translation = translator(sentence)

print("Original:", sentence)
print("Translated:", translation[0]['translation_text'])
```

```python
from transformers import pipeline

# Load translation pipeline for English to Marathi
translator = pipeline("translation_en_to_mr", model="Helsinki-NLP/opus-mt-en-mr")

# Input sentence
sentence = "How are you doing today?"

# Translate
translation = translator(sentence)

print("Original (English):", sentence)
print("Translated (Marathi):", translation[0]['translation_text'])
```

```python
from transformers import pipeline

# Load Question Answering pipeline with a popular pretrained model
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Context paragraph jisme se answer nikalna hai
context = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It is named after the engineer Gustave Eiffel, whose company designed and built the tower.
Constructed from 1887 to 1889 as the entrance to the 1889 World's Fair, it is a global cultural icon of France.
"""

# User question
question = "Who designed the Eiffel Tower?"

# Get answer
result = qa_pipeline(question=question, context=context)

print(f"Question: {question}")
print(f"Answer: {result['answer']}")
print(f"Score: {result['score']:.4f}")
```
