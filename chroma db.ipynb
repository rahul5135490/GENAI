```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection("my_collection")
```

```python
collection.add(
    documents=["मैं खुश हूँ", "मुझे दुःख है", "आज मौसम सुहाना है"],
    ids=["1", "2", "3"]
)
```

```python
results = collection.query(
    query_texts=["मुझे दुख है"],
    n_results=2
)
```

```python
print(results)
```
