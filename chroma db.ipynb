```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection("my_collection")
```

```python
collection.add(
    documents=["  ", "  ", "   "],
    ids=["1", "2", "3"]
)
```

```python
results = collection.query(
    query_texts=["  "],
    n_results=2
)
```

```python
print(results)
```
